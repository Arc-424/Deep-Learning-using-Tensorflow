```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
import numpy as np
import pandas as pd
import matplotlib
from sklearn.model_selection import train_test_split
```

```python
print(tf.__version__)
import warnings
warnings.filterwarnings('ignore')
```

```python
#click here for data shttps://www.kaggle.com/aakash50897/churn-modellingcsv
ds = pd.read_csv("Churn_Modelling.csv")
ds.head()
```

```python
x= ds.drop(labels=['RowNumber','CustomerId','Surname','Exited'],axis=1)
y= ds['Exited']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label1 = LabelEncoder()
x['Geography'] = label1.fit_transform(x['Geography'])
label1 = LabelEncoder()
x['Gender'] = label1.fit_transform(x['Gender'])
```

```python
x.head()
```

```python
x=pd.get_dummies(x,columns=['Geography'],drop_first=True)
#convert the catagorical data ,onehot encoder is also used
x.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0,stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
#build Ann
model = Sequential()
model.add(Dense(x.shape[1],activation='relu',input_dim=x.shape[1])) 
model.add(Dense(128,activation='relu')) # for the i/p layer 128 are 
model.add(Dense(1,activation='sigmoid')) # for the o/p
```

```python
x.shape[1] #if loss funtion is not decrese thenn the model is no overfit
```

```python
#compile the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train.to_numpy(),batch_size=10,epochs=10,verbose=1)
```

```python
y_pred = model.predict_classes(x_test)
y_pred
```

```python
y_test
```

```python
model.evaluate(x_test,y_test.to_numpy())
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
